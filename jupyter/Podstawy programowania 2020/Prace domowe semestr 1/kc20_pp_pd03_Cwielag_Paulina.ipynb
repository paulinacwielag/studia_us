### Zadanie 1 [20 pkt] Sekwencje
Stwórz

* [3 pkt] listę pierwiastków kwadratowych liczb od 0.0001 do 3.1415 co 0.0001
* [3 pkt] krotkę liczb parzystych od -20 do 44
* [3 pkt] ciąg znaków odwrotności nieparzystych liczb naturalnych od 1 do 99
    $$s = \sum_{j = 1}^{99} \frac{1}{j}$$
* [3 pkt] zbiór (set) wszystkich liczb od -333 do 21345 podzielnych przez 8 i 11 
* [3 pkt] listę znaków wszystkich liczb całkowitych od -1000 do 0, w których znajdziemy cyfrę "3" (proszę zapoznać się/przypomnieć sobie o funkcji rzutującej `str()`)

Oblicz sumy liczb zmagazynowanych w powyższych zbiorach danych [5 $\times$ 1 pkt].

```python
from math import sqrt
```

```python
# lista pierwiastków kwadratowych liczb od 0.0001 do 3.1415 co 0.0001

lista = []
start = 0.0001
stop = 3.1415

while start <= stop:
    lista += [sqrt(start)]
    start += 0.0001
    
print(lista)
```

```python
# krotka liczb parzystych od -20 do 44

krotka = ()
start = -20
stop = 44

while start <= stop:
    if start % 2 == 0:
        krotka += (start, )
    start += 1

print(krotka)
```

```python
# ciąg znaków odwrotności nieparzystych liczb naturalnych od 1 do 99

ciag_znakow = ""
start = 1
stop = 99

while start <= 99:
    if start % 2 == 1:
        ciag_znakow += str(1 / start) + ", "
    start += 1
    
print(ciag_znakow)
```

```python
# zbiór (set) wszystkich liczb podzielnych przez 8 i 11 od -333 do 21345

zbior = set()
start = -333
stop = 21345

while start <= stop:
    if (start % 8 == 0) and (start % 11 == 0):
        zbior.update([start])
    start += 1
    
print(zbior) 
```

```python
# lista wszystkich liczb całkowitych od -1000 do 0, w których znajdziemy cyfrę "3"

lista = []
start = -1000

while start <= 0:
    if (str(start).count("3")):
        lista += [start]
    start += 1

print(lista)
```

### Zadanie 2 Znów liczba urodzenia [20 pkt]
Za pomocą pętli (nie ręcznie) oblicz swoją [liczbę urodzenia](https://pl.wikipedia.org/wiki/Numerologia).

```python
# rozwiązanie tutaj

data = input(str("Twoja data urodzenia (dd-mm-rrrr): "))

start = 0
suma = 0

while start < len(data):
    if data[start].isdigit():
        suma += int(data[start])
    start += 1

while suma >= 10:
    data = str(suma)
    suma = int(data[0]) + int(data[1])
else:
    print(suma)
```

### Zadanie 3 Dobre hasło [20 pkt]
Większość serwisów oferuje obecnie tworzenie kont, do których to trzeba podać relatywnie silnie hasło, takie na które składa się 
* przynajmniej 8 znaków, w tym:
* przynajmniej 1 duża litera
* przynajmniej 1 cyfra
* przynajmniej 1 znak specjalny (`!@#%^&*()_+}{;:.,/?`)

Napisz program, który sprawdza czy hasło podane w zmiennej `haslo` spełnia powyższe założenia. Jeżeli spełnia to proszę poprawnie sformatować informację przekazaną funkcją `print`

    'haslo <hedh7@jA> jest poprawne'
    
lub

    'haslo <hedh7@ja> jest niepoprawne'

```python
# rozwiązanie tutaj
```

```python
haslo = input(str("Wpisz przykładowe hasło: "))

duze_litery = 0
cyfry = 0
znaki_specjalne = 0
tablica_znakow = ['!', '@', '#', '%', '^', '&', '*', '(', ')', '_', '+', '}', '{', ';', ':', '.', ',', '/', '?']

for znak in haslo:
    if znak.isupper():
        duze_litery += 1
    if znak.isdigit():
        cyfry += 1
    if znak in tablica_znakow:
        znaki_specjalne += 1
if (len(haslo) >= 8) and (duze_litery >= 1) and (cyfry >= 1) and (znaki_specjalne >= 1):
    print(f'haslo <{haslo}> jest poprawne')
else:
    print(f'haslo <{haslo}> jest niepoprawne')
```

### Zadanie 4 Generuj dobre hasło [20 pkt]
Wygeneruj hasło spełniające założenia z zadania 3.

```python
# rozwiązanie tutaj

import string
import random

haslo = ""

znaki_hasla = []

tablica_znakow = ['!', '@', '#', '%', '^', '&', '*', '(', ')', '_', '+', '}', '{', ';', ':', '.', ',', '/', '?']

# do listy zawierającej znaki hasła dodajemy kolejno 1 losową cyfrę, 
# 1 dużą literę alfabetu, 5 dowolnych liter oraz 1 znak specjalny z tablicy znaków

znaki_hasla += str(random.randint(0, 9))
znaki_hasla += random.sample(string.ascii_letters.upper(), 1)
znaki_hasla += random.sample(string.ascii_letters, 5)
znaki_hasla += random.sample(tablica_znakow, 1)

# elementy hasła dowolnie się mieszają

random.shuffle(znaki_hasla)

# za pomocą pętli for wszystkie elementy tablicy zapisujemy do ciągu znaków

for znak in znaki_hasla:
    haslo += znak
print(f"Twoje hasło to: {haslo}")
```

### Zadanie 5 [20 pkt] 1089
Proszę napisać program który dla konkretnej trzycyfrowej `ABC` (np: `543`) liczby oblicza po kolei

1. odwraca cyfry liczby oryginalnej `CBA` (`345`)

2. odejmuje liczbę odwróconą od oryginalnej `XYZ = ABC - CBA (543 - 345 = 198)`

3. odwracamy cyfry obliczonej różnicy `ZYX (891)`

4. dodajemy do siebie różnicę (`XYZ`) oraz jej odwrócenie (`ZYX`): 
`wynik = XYZ + ZYX  (198 + 891 = 1089)`

5. drukujemy powyższy `wynik`

Proszę posprawdzać taki algorytm dla kilku liczb. Jeżeli cyfry w oryginalnej liczbie ABC będą ułożone malejąco (jak w powyższym przykładzie), to wynik obliczeń będzie wynosił zawsze `1089`.

```python
# rozwiązanie tutaj
```

```python
# z treści zadania zrozumiałam, iż cyfry zawarte w liczbie trzycyfrowej muszą być od siebie różne,
# to znaczy ABC, a nie ABA, CCA itp.

liczba_trzycyfrowa = input(str("Wpisz liczbę trzycyfrową: "))

while len(liczba_trzycyfrowa) != 3:
    liczba_trzycyfrowa = input(str("BŁĄD! Wpisz liczbę trzycyfrową: "))
    
print(f"Wpisana liczba trzycyfrowa: {liczba_trzycyfrowa}")

# używam metod dla stringów 

odwrocona_liczba = str(liczba_trzycyfrowa)[::-1]
print(f"Podana liczba po jej odwróceniu: {odwrocona_liczba}")

roznica = abs(int(liczba_trzycyfrowa) - int(odwrocona_liczba))
print(f"Różnica podanej liczby trzycyfrowej oraz jej odwrócenia: {roznica}")

odwrocona_roznica = str(roznica)[::-1]
print(f"Odwrócona różnica: {odwrocona_roznica}")

wynik = int(odwrocona_roznica) + int(roznica)
print(f"Wynik: {wynik}")
```

```python
# przystosowuję program do obliczenia naraz kilku konkretnych liczb z listy, by sprawdzić jego działanie

liczby = [456, 897, 123, 943, 489]
for liczba in liczby:
    liczba_trzycyfrowa = str(liczba)

    print(f"Wpisana liczba trzycyfrowa: {liczba_trzycyfrowa}")

    odwrocona_liczba = str(liczba_trzycyfrowa)[::-1]
    print(f"Podana liczba po jej odwróceniu: {odwrocona_liczba}")

    roznica = abs(int(liczba_trzycyfrowa) - int(odwrocona_liczba))
    print(f"Różnica podanej liczby trzycyfrowej oraz jej odwrócenia: {roznica}")

    odwrocona_roznica = str(roznica)[::-1]
    print(f"Odwrócona różnica: {odwrocona_roznica}")

    wynik = int(odwrocona_roznica) + int(roznica)
    print(f"Wynik: {wynik} \n")
```
