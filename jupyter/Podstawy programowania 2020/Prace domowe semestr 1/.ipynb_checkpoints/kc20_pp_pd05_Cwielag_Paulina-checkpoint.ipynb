```python
# 1

numer_albumu = {
            "Cwielag": 65748392,
            "Dworak": 55674893,
            "Fraczkiewicz": 98763456,
            "Gorbaszewicz": 81726354,
            "Grymel": 83926574,
            "Hajduczek": 12398765,
            "Herman": 89876589,
            "Jargiło": 56473829,
            "Juraszek": 54638297,
            "Konowalow": 98274910,
            "Kowalik": 94050253,
            "Nowrotek": 11286734,
            "Lisowski": 84730097
            } 

for numer in numer_albumu:
    if numer_albumu[numer] % 2 == 0:
        numer_albumu[numer] = numer[0:4], "M"
    else:
        numer_albumu[numer] = numer[0:4], "K"
        
numer_albumu
```

```python
# 2
# Rok nazywamy przestępnym, jeżeli potraktowany jako liczba całkowita
# jest podzielny przez 4 i nie jest podzielny przez 100, lub jest 
# podzielny przez 400.

rok_urodzenia = {
            "Cwielag": 2001,
            "Dworak": 2000,
            "Fraczkiewicz": 1992,
            "Gorbaszewicz": 2001,
            "Grymel": 2001,
            "Hajduczek": 1997,
            "Herman": 2001,
            "Jargiło": 1998,
            "Juraszek": 1997,
            "Konowalow": 2001,
            "Kowalik": 2000,
            "Nowrotek": 1996,
            "Lisowski": 2001
            }

for rok in rok_urodzenia:
    if ((rok_urodzenia[rok] % 4 == 0) and (rok_urodzenia[rok] % 100 != 0)) or (rok_urodzenia[rok] % 400 == 0):
        rok_urodzenia[rok] = rok_urodzenia[rok], "jest przestepny"
    else:
        rok_urodzenia[rok] = rok_urodzenia[rok], "nie jest przestepny"
        
rok_urodzenia
```

* na podstawie danych z tego słownika utwórz drugi słownik zawierajacy dane osoby oraz jej wiek (podany w latach lub w miesiącach jeżeli wynik odnosi się do dziecka, które nie ukończyło roku).
* wypisz osoby, które mają okrągłą rocznicę urodzin (10, 20, 30, 40...)

```python
data_urodzenia = {
        ("Jan", "Kowalski"): "13.04.1965",
        ("Adam", "Nowak"): "05.11.1930",
        ("Justyna", "Święs"): "24.09.2020",
        ("Antonina", "Narta"): "12.08.2005",
        ("Kamila", "Rudowicz"): "09.12.2019",
        ("Tomasz", "Dubeltówka"): "17.12.1982",
        ("Gabriel", "Wykidajło"): "29.07.1943",
        ("Serafina", "Kotarska"): "22.11.2020",
        ("Dariusz", "Poncewicz"): "31.05.1999",
        ("Marcelina", "Musioł"): "18.09.1970",
        ("Paweł", "Leonowicz"): "09.03.2003",
        ("Grażyna", "Igiełka"): "01.01.2020"
        }

wprowadzony_rok = int(input("Podaj rok urodzenia: "))
                 
for osoba in data_urodzenia:
    rok_urodzenia = int(data_urodzenia[osoba][-4:])
    if (wprowadzony_rok > rok_urodzenia):
        print(f"{osoba[0]} {osoba[1]} urodził/a się przed rokiem {wprowadzony_rok}")
    elif (wprowadzony_rok == rok_urodzenia):
        print(f"{osoba[0]} {osoba[1]} urodził/a się w roku {wprowadzony_rok}")
    else:
        print(f"{osoba[0]} {osoba[1]} urodził/a się po roku {wprowadzony_rok}")


slownik_wiek = {}        
        
for osoba in data_urodzenia:
    rok = 2020 - int(data_urodzenia[osoba][-4:])
    miesiace = 12 - int(data_urodzenia[osoba][3:5])
    if (rok == 1):
        wiek = "1 rok"
    elif (rok > 0):
        wiek = str(rok) + " lat"
    elif miesiace > 4:
        wiek = str(miesiace) + " miesięcy"
    elif (miesiace < 4) and (miesiace > 1):
        wiek = str(miesiace) + " miesiące"
    else:
        wiek = str(miesiace) + " miesiąc"
    
    slownik_wiek[osoba[1]] = (osoba[0], osoba[1], data_urodzenia[osoba], wiek)

print(slownik_wiek)

print("Osoby obchodzące okrągłą rocznicę urodzin: ")
for osoba in data_urodzenia:
    rok = 2020 - int(data_urodzenia[osoba][-4:])
    if (rok % 10 == 0) and (rok > 0):
        print(f"{osoba[0]} {osoba[1]}, {data_urodzenia[osoba]}, {slownik_wiek[osoba[1]][3]}")
        
```
