### Zadanie 1 [30 pkt] Sekwencje i ich metody
Na początku zadania stwórz pustą zmienną sekwencyjną o podanym typie. Następnie za pomocą wskazanej metody rozszerz zmienne o kolejne elementy

* [4 pkt] Lista i metoda `append`. Dodaj pierwiastki kwadratowe liczb od -$\pi$ do $\pi$ co $\pi/100$.
* [4 pkt] Lista i metoda `extend`. Dodaj liczby parzyste i podzielne przez 7 od -100 do 444.
* [4 pkt] Zbiór i metoda `add`. Dodaj odwrotności nieparzystych liczb naturalnych od -100 do 299, tych które mają cyfrę `1` na 3 miejscu rozwinięcia binarnego.
* [4 pkt] Zbiór i metoda `update`. Dodaj wszystkie liczby od -333 do 21345 które mają literę `b` w rozwinięciu szesnastkowym.  
* [4 pkt] Słownik i metoda `update`. Dodaj pierwiastki trzeciego stopnia liczb całkowite od 1000 do 0, w których znajdziemy cyfrę `7` w rozwinięciu ósemkowym.
* [4 pkt] Słownik i standardowy interfejs `d[klucz] = wartosc`. Dodaj wszystkie liczb naturalne z zakresu od `A` do `B`, w których znajdziemy cyfrę `C`.

Z pomocą osobnej pętli `for` oblicz sumy liczb zmagazynowanych w powyższych zbiorach danych [6 $\times$ 1 pkt].

```python
# Lista i metoda append. Dodaj pierwiastki kwadratowe liczb od - 𝜋  do  𝜋  co  𝜋/100 .

from math import pi
from math import sqrt

lista1 = []

start = -pi
stop = pi

while start <= stop:
    pierwiastki_kwadratowe = sqrt(abs(start))
    lista1.append(pierwiastki_kwadratowe)
    start += pi / 100

print(lista1)
```

```python
# Lista i metoda extend. Dodaj liczby parzyste i podzielne przez 7 od -100 do 444.

lista2 = []

start = -100
stop = 444

while start <= stop:
    if (start % 7 == 0) and (start % 2 == 0):
        lista2.extend([start])
    start += 1
    
print(lista2)
```

```python
# Zbiór i metoda add. Dodaj odwrotności nieparzystych liczb naturalnych 
# od -100 do 299, tych które mają cyfrę 1 na 3 miejscu rozwinięcia 
# binarnego.

zbior1 = set()

start = -100
stop = 299

while start <= stop:
    liczba_binarnie = bin(start)
    if (start % 2 != 0) and (len(liczba_binarnie) - (liczba_binarnie.index('b') + 1) >= 3) and (liczba_binarnie[liczba_binarnie.index('b') + 3] == '1'):
        zbior1.add(1 / start)
    start += 1
    
print(zbior1)
```

```python
# Zbiór i metoda update. Dodaj wszystkie liczby od -333 do 21345 które mają literę b w rozwinięciu szesnastkowym.

zbior2 = set()

start = -333
stop = 21345

while start <= stop:
    if hex(start).count('b'):
        zbior2.update([start])
    start += 1
        
print(zbior2)
```

```python
# Słownik i metoda update. 
# Dodaj pierwiastki trzeciego stopnia liczb całkowite od 1000 do 0, w których znajdziemy cyfrę 7 w rozwinięciu ósemkowym.

d1 = {}

start = 1000
stop = 0

while start >= stop:
    if oct(start).count('7'):
        d1.update({start: start ** (1 / 3)})
    start -= 1
        
print(d1)
```

```python
# Słownik i standardowy interfejs d[klucz] = wartosc. 
# Dodaj wszystkie liczby naturalne z zakresu od A do B, w których znajdziemy cyfrę C.

# Zakres oraz cyfra C nie zostały podane, tak więc podstawiam swoje.

d2 = {}

A = 1
B = 100
C = 9

while A <= B:
    if str(A).count(str(C)):
        d2[A] = A
    A += 1
    
print(d2)
```

```python
# Z pomocą osobnej pętli for oblicz sumy liczb zmagazynowanych w powyższych zbiorach danych.


# Tworzę krotkę, w którą wpisuję wszystkie utworzone zbiory danych.

zbiory = (lista1, lista2, zbior1, zbior2, d1, d2)
print(zbiory)

# Tworzę pętlę w pętli.

suma = 0
zsumowane_liczby_zbioru = 0
kolejnosc_zbioru = 1

for zbior in zbiory:
    for element in zbior:
        suma += element
        zsumowane_liczby_zbioru += element
    print(f'Suma liczb zmagazynowanych w zbiorze {kolejnosc_zbioru} wynosi {zsumowane_liczby_zbioru}.')
    zsumowane_liczby_zbioru = 0
    kolejnosc_zbioru += 1
    
print(f'Suma liczb zmagazynowanych we wszystkich utworzonych zbiorach wynosi {suma}.')
```

```python
# rozwiązania tutaj

# zostały tylko liczby parzyste, użyj pop

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

# drukuję wylosowane liczby
print(f'Wylosowany zbiór liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if kopia_listy[start] % 2 != 0:
        kopia_listy.pop(start)
    else:
        start += 1
        
# drukuję tylko liczby parzyste z wylosowanych liczb
print(f'\nLiczby parzyste: {kopia_listy}')
```

```python
# zostały tylko liczby nieparzyste, użyj remove

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

print(f'Wylosowany zbiór liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if kopia_listy[start] % 2 == 0:
        kopia_listy.remove(kopia_listy[start])
    else:
        start += 1
        
print(f'\nLiczby nieparzyste: {kopia_listy}')
```

```python
# zostały tylko liczby mniejsze niż (B - A)/2, użyj pop

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

print(f'Wylosowany zbiór liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if kopia_listy[start] > (B - A) / 2:
        kopia_listy.pop(start)
    else:
        start += 1

print(f'\nLiczby mniejsze od {(B - A) / 2} w tym zbiorze to: {kopia_listy}')
```

```python
# w miejscu liczb podzielnych przez 3 i 7 pojawiło się None, użyj insert oraz pop !!!

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

print(f'Wylosowany zbiór liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if (kopia_listy[start] % 3 == 0) and (kopia_listy[start] % 7 == 0):
        kopia_listy.pop(start)
        kopia_listy.insert(start, None)
        start += 1
    else:
        start += 1

print(f'\nW miejscu liczb podzielnych przez 3 i 7 pojawia się None: {kopia_listy}')
```

```python
# w miejscu liczb podzielnych przez 5 i 6 pojawiło się None, użyj standardowego interfejsu do list lista[indeks] = wartosc!!!

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

print(f'Wylosowany zbiór liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if (kopia_listy[start] % 5 == 0) and (kopia_listy[start] % 6 == 0):
        kopia_listy[start] = None
        start += 1
    else:
        start += 1

print(f'\nW miejscu liczb podzielnych przez 5 i 6 pojawia się None: {kopia_listy}')
```

### Zadanie 3 [20 pkt] Zaimplementuj algorytm znajdywania wszystkich miejsc wystapienia elementu w sekwencji

1. czy element istnieje
`sekwencja.count(element)`
2. jeżeli
    * `== 0` -> koniec programu
    * `== 1` -> wystarczy `sekwencja.index(element)`
    * `> 1` -> punkt 3 
3. `> 1`
    * (3.1) zlokalizuj 1 wystapienie i zapisz ten indeks (`idx = sekwencja.index(element)`)
    * (3.2) wytnij krotkę od indeksu idx + 1 do końca i wróć do punktu 3.1
    `idx2 = sekwencja[idx + 1:].index(element) + idx + 1`

```python
# rozwiązanie tutaj

sekwencja = (3, 5 , 9, 10, 3, 9, 10, 10, 5, 9, 5, 5)

element = int(input("Podaj element: "))

if (sekwencja.count(element) == 0):
    print("Element nie występuje w sekwencji.")
elif (sekwencja.count(element) == 1):
    print(f"Element występuje w sekwencji o indeksie {sekwencja.index(element)}")
else:
    print(f"Element {element} występuje w krotce {sekwencja.count(element)} razy.")
    idx = sekwencja[0:].index(element)
while idx < len(sekwencja):
        if sekwencja[idx + 1:].count(element) == 0:
            break
        idx2 = sekwencja[idx + 1:].index(element) + (idx + 1)
        idx = idx2
        print(f"Element {element} występuje w krotce na miejscu o indeksie {idx}.")
```

### Wstęp do zadania 5
Słowniki zwykle służą do tłumaczenia słów z jednego języka na drugi. Tak samo można wykorzystać słowniki pythonowskie. Zobaczcie ten snippet (da się ładniej, ale to kiedy indziej...)

```python
d = {'kot': 'cat', 'pies': 'dog'}

zdanie = 'kot i pies w jednym stali domu'
zdanie_en = ''

for slowo in zdanie.split(' '):
    if slowo in d:
        zdanie_en += d[slowo]
    else:
        zdanie_en += slowo
    zdanie_en += ' '

print(zdanie_en)
```
