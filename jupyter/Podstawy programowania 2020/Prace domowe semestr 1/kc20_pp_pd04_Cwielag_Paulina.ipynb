### Zadanie 1 [30 pkt] Sekwencje i ich metody
Na pocztku zadania stw贸rz pust zmienn sekwencyjn o podanym typie. Nastpnie za pomoc wskazanej metody rozszerz zmienne o kolejne elementy

* [4 pkt] Lista i metoda `append`. Dodaj pierwiastki kwadratowe liczb od -$\pi$ do $\pi$ co $\pi/100$.
* [4 pkt] Lista i metoda `extend`. Dodaj liczby parzyste i podzielne przez 7 od -100 do 444.
* [4 pkt] Zbi贸r i metoda `add`. Dodaj odwrotnoci nieparzystych liczb naturalnych od -100 do 299, tych kt贸re maj cyfr `1` na 3 miejscu rozwinicia binarnego.
* [4 pkt] Zbi贸r i metoda `update`. Dodaj wszystkie liczby od -333 do 21345 kt贸re maj liter `b` w rozwiniciu szesnastkowym.  
* [4 pkt] Sownik i metoda `update`. Dodaj pierwiastki trzeciego stopnia liczb cakowite od 1000 do 0, w kt贸rych znajdziemy cyfr `7` w rozwiniciu 贸semkowym.
* [4 pkt] Sownik i standardowy interfejs `d[klucz] = wartosc`. Dodaj wszystkie liczb naturalne z zakresu od `A` do `B`, w kt贸rych znajdziemy cyfr `C`.

Z pomoc osobnej ptli `for` oblicz sumy liczb zmagazynowanych w powy偶szych zbiorach danych [6 $\times$ 1 pkt].

```python
# rozwizania tutaj
```

```python
# Lista i metoda append. Dodaj pierwiastki kwadratowe liczb od -   do    co  /100 .

from math import pi
from math import sqrt

lista1 = []

start = -pi
stop = pi

while start <= stop:
    pierwiastki_kwadratowe = sqrt(abs(start))
    lista1.append(pierwiastki_kwadratowe)
    start += pi / 100

print(lista1)
```

```python
# Lista i metoda extend. Dodaj liczby parzyste i podzielne przez 7 od -100 do 444.

lista2 = []

start = -100
stop = 444

while start <= stop:
    if (start % 7 == 0) and (start % 2 == 0):
        lista2.extend([start])
    start += 1
    
print(lista2)
```

```python
# Zbi贸r i metoda add. Dodaj odwrotnoci nieparzystych liczb naturalnych 
# od -100 do 299, tych kt贸re maj cyfr 1 na 3 miejscu rozwinicia 
# binarnego.

zbior1 = set()

start = -100
stop = 299

while start <= stop:
    liczba_binarnie = bin(start)
    if (start % 2 != 0) and (len(liczba_binarnie) - (liczba_binarnie.index('b') + 1) >= 3) and (liczba_binarnie[liczba_binarnie.index('b') + 3] == '1'):
        zbior1.add(1 / start)
    start += 1
    
print(zbior1)
```

```python
# Zbi贸r i metoda update. Dodaj wszystkie liczby od -333 do 21345 kt贸re maj liter b w rozwiniciu szesnastkowym.

zbior2 = set()

start = -333
stop = 21345

while start <= stop:
    if hex(start).count('b'):
        zbior2.update([start])
    start += 1
        
print(zbior2)
```

```python
# Sownik i metoda update. 
# Dodaj pierwiastki trzeciego stopnia liczb cakowite od 1000 do 0, w kt贸rych znajdziemy cyfr 7 w rozwiniciu 贸semkowym.

d1 = {}

start = 1000
stop = 0

while start >= stop:
    if oct(start).count('7'):
        d1.update({start: start ** (1 / 3)})
    start -= 1
        
print(d1)
```

```python
# Sownik i standardowy interfejs d[klucz] = wartosc. 
# Dodaj wszystkie liczby naturalne z zakresu od A do B, w kt贸rych znajdziemy cyfr C.

# Zakres oraz cyfra C nie zostay podane, tak wic podstawiam swoje.

d2 = {}

A = 1
B = 100
C = 9

while A <= B:
    if str(A).count(str(C)):
        d2[A] = A
    A += 1
    
print(d2)
```

```python
# Z pomoc osobnej ptli for oblicz sumy liczb zmagazynowanych w powy偶szych zbiorach danych.


# Tworz krotk, w kt贸r wpisuj wszystkie utworzone zbiory danych.

zbiory = (lista1, lista2, zbior1, zbior2, d1, d2)
print(zbiory)

# Tworz ptl w ptli.

suma = 0
zsumowane_liczby_zbioru = 0
kolejnosc_zbioru = 1

for zbior in zbiory:
    for element in zbior:
        suma += element
        zsumowane_liczby_zbioru += element
    print(f'Suma liczb zmagazynowanych w zbiorze {kolejnosc_zbioru} wynosi {zsumowane_liczby_zbioru}.')
    zsumowane_liczby_zbioru = 0
    kolejnosc_zbioru += 1
    
print(f'Suma liczb zmagazynowanych we wszystkich utworzonych zbiorach wynosi {suma}.')
```

```python
# rozwizania tutaj

# zostay tylko liczby parzyste, u偶yj pop

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

# drukuj wylosowane liczby
print(f'Wylosowany zbi贸r liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if kopia_listy[start] % 2 != 0:
        kopia_listy.pop(start)
    else:
        start += 1
        
# drukuj tylko liczby parzyste z wylosowanych liczb
print(f'\nLiczby parzyste: {kopia_listy}')
```

```python
# zostay tylko liczby nieparzyste, u偶yj remove

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

print(f'Wylosowany zbi贸r liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if kopia_listy[start] % 2 == 0:
        kopia_listy.remove(kopia_listy[start])
    else:
        start += 1
        
print(f'\nLiczby nieparzyste: {kopia_listy}')
```

```python
# zostay tylko liczby mniejsze ni偶 (B - A)/2, u偶yj pop

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

print(f'Wylosowany zbi贸r liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if kopia_listy[start] > (B - A) / 2:
        kopia_listy.pop(start)
    else:
        start += 1

print(f'\nLiczby mniejsze od {(B - A) / 2} w tym zbiorze to: {kopia_listy}')
```

```python
# w miejscu liczb podzielnych przez 3 i 7 pojawio si None, u偶yj insert oraz pop !!!

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

print(f'Wylosowany zbi贸r liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if (kopia_listy[start] % 3 == 0) and (kopia_listy[start] % 7 == 0):
        kopia_listy.pop(start)
        kopia_listy.insert(start, None)
        start += 1
    else:
        start += 1

print(f'\nW miejscu liczb podzielnych przez 3 i 7 pojawia si None: {kopia_listy}')
```

```python
# w miejscu liczb podzielnych przez 5 i 6 pojawio si None, u偶yj standardowego interfejsu do list lista[indeks] = wartosc!!!

from random import randint
N = 333
A = 13
B = 1013
liczby = [randint(A, B) for i in range(N)]

kopia_listy = liczby.copy()

print(f'Wylosowany zbi贸r liczb to: {liczby}')

start = 0

while start < len(kopia_listy):
    if (kopia_listy[start] % 5 == 0) and (kopia_listy[start] % 6 == 0):
        kopia_listy[start] = None
        start += 1
    else:
        start += 1

print(f'\nW miejscu liczb podzielnych przez 5 i 6 pojawia si None: {kopia_listy}')
```

### Zadanie 3 [20 pkt] Zaimplementuj algorytm znajdywania wszystkich miejsc wystapienia elementu w sekwencji

1. czy element istnieje
`sekwencja.count(element)`
2. je偶eli
    * `== 0` -> koniec programu
    * `== 1` -> wystarczy `sekwencja.index(element)`
    * `> 1` -> punkt 3 
3. `> 1`
    * (3.1) zlokalizuj 1 wystapienie i zapisz ten indeks (`idx = sekwencja.index(element)`)
    * (3.2) wytnij krotk od indeksu idx + 1 do koca i wr贸 do punktu 3.1
    `idx2 = sekwencja[idx + 1:].index(element) + idx + 1`

```python
# rozwizanie tutaj

sekwencja = (3, 5 , 9, 10, 3, 9, 10, 10, 5, 9, 5, 5)

element = int(input("Podaj element: "))

if (sekwencja.count(element) == 0):
    print("Element nie wystpuje w sekwencji.")
elif (sekwencja.count(element) == 1):
    print(f"Element wystpuje w sekwencji o indeksie {sekwencja.index(element)}")
else:
    print(f"Element {element} wystpuje w krotce {sekwencja.count(element)} razy.")
    idx = sekwencja[0:].index(element)
while idx < len(sekwencja):
        if sekwencja[idx + 1:].count(element) == 0:
            break
        idx2 = sekwencja[idx + 1:].index(element) + (idx + 1)
        idx = idx2
        print(f"Element {element} wystpuje w krotce na miejscu o indeksie {idx}.")
```

### Wstp do zadania 5
Sowniki zwykle su偶 do tumaczenia s贸w z jednego jzyka na drugi. Tak samo mo偶na wykorzysta sowniki pythonowskie. Zobaczcie ten snippet (da si adniej, ale to kiedy indziej...)

```python
d = {'kot': 'cat', 'pies': 'dog'}

zdanie = 'kot i pies w jednym stali domu'
zdanie_en = ''

for slowo in zdanie.split(' '):
    if slowo in d:
        zdanie_en += d[slowo]
    else:
        zdanie_en += slowo
    zdanie_en += ' '

print(zdanie_en)
```
