```python
def czy_pal(x):
    """x: lista
       zwraca True jezeli lista jest palindromem
       jezeli nie zwraca False
       
    >>> czy_pal(['z', 'a', 'r', 'a', 'z'])
    True
    >>> czy_pal(['h', 'e', 'r', 'b', 'a', 't', 'a'])
    False
    >>> czy_pal(['k', 'a', 'j', 'a', 'k', '1'])
    False
    >>> czy_pal(['9', 'k', 'a', 'j', 'a', 'k', '9'])
    True
    >>> czy_pal([])
    False
    >>> czy_pal(['i'])
    False
    >>> czy_pal([1, 2, 3])
    False
    >>> czy_pal([1, 2, 1])
    True
    """
    temp = x
    temp.reverse
    if temp == x:
        return True
    else:
        return False
    
import doctest
doctest.testmod()
```

```python
import doctest

def srednia_geometryczna(L):
    '''
    Funkcja oblicza średnią geometryczną z liczb listy L.
    WE: L - lista liczb
    WY: FLOAT - średnia geometryczna liczb z listy L

    >>> srednia_geometryczna([1])
    1.0
    >>> srednia_geometryczna([2, 2, 2])
    2.0
    >>> srednia_geometryczna([1])
    1.0
    >>> srednia_geometryczna([4, 6, 8, 10])
    6.6195
    >>> srednia_geometryczna([])
    0.0
    >>> srednia_geometryczna([0])
    0.0
    >>> srednia_geometryczna([3.2, 2.4, 8.9, 5.6])
    4.42318
    '''
    from math import sqrt
    srednia = 1
    i = 0
    while L:
        srednia = srednia * L.pop()
        i += 1
    if i != 0:
        return round(srednia ** (1 / i), 5)
    else:
        return 0.0


doctest.testmod()


def test_srednia_geometryczna():

    # 1: Lista conajmniej jednoelementowa
    assert srednia_geometryczna([1, 2, 3]) == 1.81712, "srednia_geometryczna([1, 2, 3]) != 1.817"
    # 2: Lista pusta
    assert srednia_geometryczna([]) == 0.0, "srednia_geometryczna([]) != 0.0"

test_srednia_geometryczna()
```
