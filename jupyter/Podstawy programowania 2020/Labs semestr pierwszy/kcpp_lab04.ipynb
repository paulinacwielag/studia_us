```python
liczba1 = float(input("Podaj liczbę 1: "))
liczba2 = float(input("Podaj liczbę 2: "))

print("1: +, 2: -, 3: *, 4: /, 5: //, 6: %, 7: sqrt, 8: KONIEC")

menu = int(input("Co chcesz zrobić: "))

if menu == 1:
    print(f"Suma liczb {liczba1} i {liczba2} jest równa {liczba1 + liczba2}")
elif menu == 2:
    print(liczba1 - liczba2)
elif menu == 3:
    print(liczba1 * liczba2)
elif menu == 4:
    if liczba2 == 0:
        print("Nie dzielimy przez 0.")
    else:
        print(f"Iloraz liczb: {liczba1} i {Liczba2} wynosi {liczba1 / liczba2}.")
elif menu == 5:
    print(liczba1 // liczba2)
elif menu == 6:
    print(liczba1 % liczba2)
elif menu == 7:
    print(liczba1 ** (1 / liczba2))
else:
    print("Koniec programu")
```
