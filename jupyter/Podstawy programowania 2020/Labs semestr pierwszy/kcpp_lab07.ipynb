```python
zbior = set()
start = 100
stop = -100
suma = 0

i = start
while i >= stop:
    if start % 2 == 0:
        zbior.add(i ** 2)
    i -= 1
    
print(zbior)

suma = 0
for element in zbior:
    suma + (element ** 0.5)

print(f"Suma:", suma)
```

```python
zbior = set()
start = 100
stop = -100

# algorytm
i = start
while i >= stop:
    if i % 2 == 0: 
        zbior.add(i ** 2)
    i -= 1

print(zbior)

suma = 0
for element in zbior:
    suma += (element ** 0.5)

print("Suma:", suma)

iloczyn = 1
for element in zbior:
    iloczyn *= (element ** 0.5)
    
print("Iloczyn:", iloczyn)
```

```python
slownik_kwadratow = {}
iloczyn_kluczy = 1
suma = 0
start = 0
stop = -100

while start >= stop:
    if start % 2 == 1:
        kwadrat = start ** 2
        slownik_kwadratow.update({start: kwadrat})
        suma += kwadrat
        iloczyn_kluczy *= start
    start -= 1

print("Suma kwadratów liczb nieparzystych: {}, a iloczyn kluczy to: {}".format(suma, iloczyn_kluczy))
```
