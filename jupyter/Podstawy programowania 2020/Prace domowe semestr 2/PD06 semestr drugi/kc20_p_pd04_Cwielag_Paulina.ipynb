    >>> w = Wektor(0, 0, 1, 1)

    zaczepiony w punkcie (0, 0) i zorientownay pod katem 0.785398163397 do osi x.

```python
class Wektor:
    '''Klasa Wektor'''
    
    def __init__(self, zaczepienie_x, zaczepienie_y, koncowy_x, koncowy_y):
        self.zaczepienie_x = zaczepienie_x
        self.zaczepienie_y = zaczepienie_y
        self.koncowy_x = koncowy_x
        self.koncowy_y = koncowy_y
        
    def dl_wektora(self):
        zaczepienie_x = self.zaczepienie_x
        zaczepienie_y = self.zaczepienie_y
        koncowy_x = self.koncowy_x
        koncowy_y = self.koncowy_y
        
        dl_wektora = ((koncowy_x - zaczepienie_x) ** 2 + (koncowy_y - zaczepienie_y) ** 2) ** (1/2)
        
        return dl_wektora
        
    def nachylenie_ox(self):
        import math
        
        zaczepienie_x = self.zaczepienie_x
        zaczepienie_y = self.zaczepienie_y
        koncowy_x = self.koncowy_x
        koncowy_y = self.koncowy_y
        
        bok_a = abs(koncowy_x - zaczepienie_x)
        bok_b = abs(koncowy_y - zaczepienie_y)
        alfa = math.degrees(math.atan(bok_b/bok_a))
        
        return alfa
        
    def nachylenie_oy(self):
        import math
        
        zaczepienie_x = self.zaczepienie_x
        zaczepienie_y = self.zaczepienie_y
        koncowy_x = self.koncowy_x
        koncowy_y = self.koncowy_y
        
        bok_a = abs(koncowy_x - zaczepienie_x)
        bok_b = abs(koncowy_y - zaczepienie_y)
        alfa = 90 - math.degrees(math.atan(bok_b/bok_a))
        
        return alfa
        
    def przeciwny_zwrot(self):
        zaczepienie_x = self.zaczepienie_x
        zaczepienie_y = self.zaczepienie_y
        koncowy_x = self.koncowy_x
        koncowy_y = self.koncowy_y
        
        punkt_zaczepienia = (koncowy_x, koncowy_y)
        punkt_koncowy = (zaczepienie_x, zaczepienie_y)
        
        return f'Wektor początkowy: punkt zaczepienia ({zaczepienie_x}, {zaczepienie_y}) punkt końcowy ({koncowy_x}, {koncowy_y}) \nWektor o przeciwnym zwrocie ma punkt zaczepienia {punkt_zaczepienia} i punkt końcowy {punkt_koncowy}.'
    
    def __str__(self):
        zaczepienie_x = self.zaczepienie_x
        zaczepienie_y = self.zaczepienie_y
        koncowy_x = self.koncowy_x
        koncowy_y = self.koncowy_y
        
        ret = f'Wektor o długości {self.dl_wektora()} \n'
        ret += f'zaczepiony w punkcie ({zaczepienie_x}, {zaczepienie_y}) '
        ret += f'i zorientowany pod katem {self.nachylenie_ox()} do osi x.'
        return ret
        
    def __repr__(self):
        zaczepienie_x = self.zaczepienie_x
        zaczepienie_y = self.zaczepienie_y
        koncowy_x = self.koncowy_x
        koncowy_y = self.koncowy_y
        return f'punkt zaczepienia: ({zaczepienie_x}, {zaczepienie_y}) punkt końcowy: ({koncowy_x}, {koncowy_y})'
        
    
w1 = Wektor(0, 0, 1, 1)
w2 = Wektor(2, 3, 3, 2)
w3 = Wektor(-10, -3, 98, 101)
w4 = Wektor(-1, -2, -1, -2)
w5 = Wektor(5, 10, 15, 20)
```
