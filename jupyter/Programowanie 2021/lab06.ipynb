```python
import string
from urllib.request import urlopen, Request 

url_rj = 'https://wolnelektury.pl/media/book/txt/pruszynski-palestyna-po-raz-trzeci.txt'
r = Request(url_rj)
odp = urlopen(r)
rj = odp.read().decode('utf-8')

tekst = ''
pozytywne = ['wartościowy', 'sympatyczny']
negatywne = ['nienawidzić', 'śmierci'] 
pozy = 0
neg = 0
pozy_slowa = []
neg_slowa = []

for linijka in rj:
    tekst += linijka

for slowo in tekst.lower().split():
        if slowo in pozytywne:
            pozy += 1
            pozy_slowa.append(slowo)
        if slowo in negatywne:
            neg += 1
            neg_slowa.append(slowo)
print(pozy, neg)
print('pozytywne', pozy_slowa)
print('negatywne', neg_slowa)
```
