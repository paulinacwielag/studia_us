```python
def czy_wygrana(twoje_numery, wygrane_lotto, wygrane_lotto_plus):
    
    lotto = 0
    lotto_plus = 0
    
    for liczba in twoje_numery:
        if liczba in wygrane_lotto:
            lotto += 1
        if liczba in wygrane_lotto_plus:
            lotto_plus += 1
    if (lotto < 3) and (lotto_plus < 3):
        return False
    else:
        return True

wygrane_kupony = 0
for kupon in moje_kupony:    
    for numery in kupon:
        if czy_wygrana(numery, wygrane_lotto, wygrane_lotto_plus):
            wygrane_kupony += 1

print(f'Wygrane kupony: {wygrane_kupony}')
```
