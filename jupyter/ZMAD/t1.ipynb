```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count = CountVectorizer(ngram_range=(2,3))
docs = np.array(['The sun is shining',
'The weather is sweet',
'The sun is shining, the weather is sweet,'
'and one and one is two'])
bag = count.fit_transform(docs)
```

```python
print(count.vocabulary_)
```

```python
print(bag.toarray())
```

```python
# Define a function percent(word, text) that calculates how often a given word occurs in a text 
# and expresses the result as a percentage.
```

```python
def percent(word, text):
    '''funkcja'''
    
    procent = 100 * text.count(word) / len(text)
    return procent
```

```python
print(percent('God', text3))
```

```python
print(percent('love', text2))
```
