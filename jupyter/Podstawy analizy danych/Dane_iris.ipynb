```python
import scipy # biblioteka do obliczen statystycznych
from scipy.stats import shapiro 
shapiro(setosa['petal_length'])
```

```python
for el in setosa.columns: # poziom istotności alfa, alfa > 0.05, przyjęte z góry dopuszczalne ryzyko popełnienia błędu I rodzaju (uznania prawdziwej hipotezy zerowej za fałszywą)
    if el != 'species':
        print(shapiro(setosa[el]))
```

```python
for el in versicolor.columns:
    if el != 'species':
        print(shapiro(versicolor[el]))
```

```python
# dla zmiennych dla przyjętego poziomu istotności alfa = 0.05 przyjmujemy hipotezę o normalności rozkładu (p > 0.05)
```

```python
from scipy.stats import mannwhitneyu

mannwhitneyu(setosa['petal_width'], versicolor['petal_width'])
```

```python
from scipy.stats import mannwhitneyu

print('Setosa vs Versicolor\n')

for column in setosa:
    print(column + '\t', end='')
    print(mannwhitneyu(setosa[column],versicolor[column]))
```

```python
# testy T (nie do czegoś co nie ma rozkładu normalnego), testy Manna Whitneya
```

```python
# wstępna wizualizacja, sprawdzamy normalnosc rozkladu tych zmiennych z podzialem na grupy, porownujemy testami istotnosci odpoiwednio dla parametrzyczmych i nieparametrycznych, normalne - test t, nie noemalne - test manna whitneya
```
